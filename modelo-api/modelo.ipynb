```python
heart = pd.read_csv('heart.csv')
```

```python
heart.head()
```

```python
heart.isna().sum()
```

```python
heart['Sex'] = heart['Sex'].map({'F':0, 'M':1})
heart['ChestPainType'] = heart['ChestPainType'].map({'ASY':0, 'NAP':1, 'ATA': 2, 'TA': 3})
heart['RestingECG'] = heart['RestingECG'].map({'Normal':0, 'LVH':1, 'ST': 2})
heart['ExerciseAngina'] = heart['ExerciseAngina'].map({'N':0, 'Y':1})
heart['ST_Slope'] = heart['ST_Slope'].map({'Up':0, 'Flat':1, 'Down': 2})
```

```python
from sklearn.model_selection import train_test_split

X = heart.drop(['HeartDisease'], axis=1)
y = heart['HeartDisease']
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y,
                                                        test_size=0.2,
                                                        stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

knn.fit(X_treino, y_treino)
```

```python
y_predito = knn.predict(X_teste)
```

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score,\
                            precision_score, confusion_matrix


print('Acurácia é', accuracy_score(y_teste, y_predito))
print('F1-score é', f1_score(y_teste, y_predito))
print('Sensibilidade é', recall_score(y_teste, y_predito))
print('Precisão é', precision_score(y_teste, y_predito))
print('\n\n\n', confusion_matrix(y_teste, y_predito))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

arvore = DecisionTreeClassifier()
random_forest = RandomForestClassifier(n_estimators=200)

arvore.fit(X_treino, y_treino)
random_forest.fit(X_treino, y_treino)
```

```python
y_predito_arvore = arvore.predict(X_teste)
y_predito_rf = random_forest.predict(X_teste)
```

```python
print('Acurácia é', accuracy_score(y_teste, y_predito_arvore))
print('F1-score é', f1_score(y_teste, y_predito_arvore))
print('Sensibilidade é', recall_score(y_teste, y_predito_arvore))
print('Precisão é', precision_score(y_teste, y_predito_arvore))
print('\n\n\n', confusion_matrix(y_teste, y_predito_arvore))
```

```python
print('Acurácia é', accuracy_score(y_teste, y_predito_rf))
print('F1-score é', f1_score(y_teste, y_predito_rf))
print('Sensibilidade é', recall_score(y_teste, y_predito_rf))
print('Precisão é', precision_score(y_teste, y_predito_rf))
print('\n\n\n', confusion_matrix(y_teste, y_predito_rf))
```
